```python
%matplotlib inline

import librosa
import librosa.display
import matplotlib.pyplot as plt

start = 100000
data,_ = librosa.load("./dog/dog_barking_{}.wav".format(6))
stft = librosa.stft(data[start:start + 60000])


plt.rcParams["figure.figsize"] = (20,20)
#plt.subplot(112, 1, i)
librosa.display.specshow(stft)
  
```

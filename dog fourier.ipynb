```python
#%matplotlib inline
import IPython.display as ipd
import librosa
import librosa.display
import matplotlib.pyplot as plt
import matplotlib

start = 100
data, _ = librosa.load("./dog/dog_barking_{}.wav".format(11))
stft = librosa.stft(data[start:start + 60000])


plt.figure(figsize=(30, 30))
librosa.display.specshow(stft, y_axis='log', x_axis='time')
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 50}

matplotlib.rc('font', **font)
plt.colorbar(format='%+2.0f dB')
plt.title('dog')
plt.tight_layout()
ipd.Audio(data, rate=_) 
```
